```python
import pandas as pd

df_sample = pd.read_csv('1000 reviews with normalized scores.csv')
#df_sample.columns.values
df_sample = df_sample.rename(columns = {'Unnamed: 0':'Index'}).set_index('Index')
df_sample.head()
```

```python
df_full_reviews = pd.read_json('yelp_academic_dataset_review_cleaned.json', lines = True)
df_full_sample = df_full_reviews.iloc[df_sample.index.values]
df_full_sample.head()

df_sample['text'] = df_full_sample['text']
df_sample.head()
```

```python
import re

def review_length(s):
    return len(re.split('\s',s))

df_sample['review_length'] = df_sample['text'].map(review_length)
df_sample.head()
```

```python
df_sample.to_csv('1000 reviews with normalized scores.csv')
```
