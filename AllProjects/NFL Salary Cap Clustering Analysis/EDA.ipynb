```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

## EDA

### Trends over Time

```python
Df3 = pd.read_csv('3Buckets.csv')
Df5 = pd.read_csv('5Buckets.csv')
Df8 = pd.read_csv('8Buckets.csv')
Df12 = pd.read_csv('12Buckets.csv')
```

```python
Df12.head()
```

```python
Df3['TotCap'] = Df3['Off'] + Df3['Def'] + Df3['Special']
Df5['TotCap'] = Df5['Fantasy'] + Df5['Front7'] + Df5['Dbacks'] + Df5['Oline'] + Df5['Special']
Df8['TotCap'] = Df8['Dbacks'] + Df8['LineBackers'] + Df8['Dline'] + Df8['PassCatchers'] + Df8['Oline'] + Df8['RB'] + Df8['QB'] + Df8['Special']
Df12['TotCap'] = Df12['Safeties'] + Df12['Corners'] + Df12['InsideBackers'] + Df12['Edge'] + Df12['InteriorDLine'] + Df12['WRs'] + Df12['TEs'] + Df12['InteriorOLine'] + Df12['Tackles'] + Df12['QB'] + Df12['RB'] + Df12['Special']
```

```python
plt.scatter(Df3['Year'], Df3['Off'])
z = np.polyfit(Df3['Year'],Df3['Off'],1)
p = np.poly1d(z)
plt.plot(Df3['Year'],p(Df3['Year']), "r--")
plt.title("Percent Cap Spent on Offense Increases")
```

```python
plt.scatter(Df8['Year'], Df8['PassCatchers'] + Df8['Oline'] + Df8['RB'])
z = np.polyfit(Df8['Year'], Df8['PassCatchers'] + Df8['Oline'] + Df8['RB'],1)
p = np.poly1d(z)
plt.plot(Df8['Year'],p(Df8['Year']), "r--")
plt.title("Even if you remove QB ")
```

```python
plt.scatter(Df3['Year'], Df3['Def'])
z = np.polyfit(Df3['Year'],Df3['Def'],1)
p = np.poly1d(z)
plt.plot(Df3['Year'],p(Df3['Year']), "r--")
plt.title("Percent Cap Spent on Defense Decreases")
```

```python
plt.scatter(Df12['Year'], Df12['TotCap'])
z = np.polyfit(Df12['Year'],Df12['TotCap'],1)
p = np.poly1d(z)
plt.plot(Df12['Year'],p(Df12['Year']), "r--")
plt.title("Teams use more of their Cap Space (2010 Technically Uncapped)")
```

```python
plt.scatter(Df8['Year'], Df8['Oline'])
z = np.polyfit(Df8['Year'],Df8['Oline'],1)
p = np.poly1d(z)
plt.plot(Df8['Year'],p(Df8['Year']), "r--")
plt.title("Teams Spend more on Oline")
```

```python
plt.scatter(Df8['Year'], Df8['RB'])
z = np.polyfit(Df8['Year'],Df8['RB'],1)
p = np.poly1d(z)
plt.plot(Df8['Year'],p(Df8['Year']), "r--")
plt.title("Teams Spend Less on Running Backs")
```

```python
plt.scatter(Df8['Year'], Df8['PassCatchers'])
z = np.polyfit(Df8['Year'],Df8['PassCatchers'],1)
p = np.poly1d(z)
plt.plot(Df8['Year'],p(Df8['Year']), "r--")
plt.title("Teams Spend More on Pass Catchers")
```

### Simple Wins Correlations

```python
plt.scatter(Df8['QB'], Df8['Wins'])
z = np.polyfit(Df8['QB'],Df8['Wins'],1)
p = np.poly1d(z)
plt.plot(Df8['QB'],p(Df8['QB']), "r--")
plt.title("Teams That Spend more on QBs Don't Necessarily Win More")
```

```python
plt.scatter(Df8['RB'], Df8['Wins'])
z = np.polyfit(Df8['RB'],Df8['Wins'],1)
p = np.poly1d(z)
plt.plot(Df8['RB'],p(Df8['RB']), "r--")
plt.title("Teams That Spend more on RBs Don't Necessarily Win More")
```

```python
plt.scatter(Df8['Oline'], Df8['Wins'])
z = np.polyfit(Df8['Oline'],Df8['Wins'],1)
p = np.poly1d(z)
plt.plot(Df8['Oline'],p(Df8['Oline']), "r--")
plt.title("Teams That Spend more on Oline actually win slightly fewer")
```

```python
plt.scatter(Df8['LineBackers'], Df8['Wins'])
z = np.polyfit(Df8['LineBackers'],Df8['Wins'],1)
p = np.poly1d(z)
plt.plot(Df8['LineBackers'],p(Df8['LineBackers']), "r--")
plt.title("Teams That Spend more on Linebackers actually win slightly more")
print ("y=%.6fx+(%.6f)"%(z[0],z[1]))
```
