```python
import pandas as pd
import sklearn as sk
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
Df3 = pd.read_csv('3Buckets.csv')
Df5 = pd.read_csv('5Buckets.csv')
Df8 = pd.read_csv('8Buckets.csv')
Df12 = pd.read_csv('12Buckets.csv')
```

```python
labels = ['Dbacks','Front7','Fantasy','Oline','Special']

for label in labels:
    Df5[label] = (Df5[label]-np.mean(Df5[label]))/np.std(Df5[label])
```

```python
Df5.head()
```

```python
X = Df5[['Dbacks','Front7','Fantasy','Oline','Special']]
NonPredictorX = Df5[['Year','Team']]
Y = Df5[['Wins', 'Playoff Wins', 'SuperBowl Win']]
```

```python
## Elbow Plot
models = []
wcss = []
for i in range(1,10): 
    i = int(i)
    KM = KMeans(n_clusters = i, random_state = 0, init = "k-means++", n_init = 10)
    KM.fit(X)
    wcss.append(KM.inertia_)
    
plt.plot(range(1,10),wcss)
plt.title("Elbow Plot")
plt.xlabel("Num Clusters")
plt.ylabel("Within Cluster Sum of Squares")

plt.show()
print("Seemingly the best number of clusters is 3,5, or 7")
```

## 3 Clusters

```python
## 3 clusters
KM5_3 = KMeans(n_clusters = 3, random_state = 0, init = "k-means++", n_init = 100)
KM5_3.fit(X)
Df5_3Clust = Df5.copy()
Df5_3Clust['Cluster'] = KM5_3.labels_
KM5_3.cluster_centers_
```

```python
##Match Average Offense, Defense, Special spending back to groupby to gain cluster averages for all metrics
Df5_3_GB = Df5_3Clust.groupby(by = "Cluster")[['SuperBowl Win', "Playoff Wins",'Wins']].mean()
Df5_3_GB.reset_index(inplace = True)
labs = X.columns
# for lab in labs:
#     Df5_3_GB[lab] = 0

for i in range (0,len(KM5_3.cluster_centers_)):
    k_m = 0
    for lab in labs:
        val = KM5_3.cluster_centers_[i][k_m]
        Df5_3_GB.set_value(i,lab,val)
        k_m+=1
        
Df5_3_GB['TotCap'] = Df5_3_GB['Dbacks'] + Df5_3_GB['Front7'] + Df5_3_GB['Fantasy'] + Df5_3_GB['Oline'] + Df5_3_GB['Special']
Df5_3_GB['Count'] = Df5_3Clust.groupby(by = "Cluster")['Year'].count()
Df5_3_GB
```

```python
## Calculate Average Prediction Error
for i in range(0,3):
    Df5_3Clust[i] = 0

def DistFunc(x):
    D0 = np.sqrt((x['Dbacks']-Df5_3_GB.iat[0,4])**2+(x['Front7'] - Df5_3_GB.iat[0,5])**2 + (x['Fantasy'] - Df5_3_GB.iat[0,6])**2
                +(x['Oline']-Df5_3_GB.iat[0,7])**2+(x['Special']-Df5_3_GB.iat[0,8])**2)
    
    D1 = np.sqrt((x['Dbacks']-Df5_3_GB.iat[1,4])**2+(x['Front7'] - Df5_3_GB.iat[1,5])**2 + (x['Fantasy'] - Df5_3_GB.iat[1,6])**2
                +(x['Oline']-Df5_3_GB.iat[1,7])**2+(x['Special']-Df5_3_GB.iat[1,8])**2)
    
    D2 = np.sqrt((x['Dbacks']-Df5_3_GB.iat[2,4])**2+(x['Front7'] - Df5_3_GB.iat[2,5])**2 + (x['Fantasy'] - Df5_3_GB.iat[2,6])**2
                +(x['Oline']-Df5_3_GB.iat[2,7])**2+(x['Special']-Df5_3_GB.iat[2,8])**2)
    
    Tot = D0 + D1 + D2
    
    N0 = 1/((D0/Tot))
    N1 = 1/((D1/Tot))
    N2 = 1/((D2/Tot))
    TotN = N0 + N1 + N2
    
    Perc0 = N0/TotN*100
    Perc1 = N1/TotN*100
    Perc2 = N2/TotN*100
    
    return pd.Series([Perc0,Perc1,Perc2])

def ExpectedWins(x):
    WinShares0 = x[0]/100*Df5_3_GB.iat[0,3]
    WinShares1 = x[1]/100*Df5_3_GB.iat[1,3]
    WinShares2 = x[2]/100*Df5_3_GB.iat[2,3]
    ExpWins = WinShares0 + WinShares1 + WinShares2
    return ExpWins

Df5_3Clust[[0,1,2]] = Df5_3Clust.apply(DistFunc, axis = 1)
    
Df5_3Clust['ExpWins'] = Df5_3Clust.apply(ExpectedWins, axis = 1)
Df5_3Clust['Error'] = abs(Df5_3Clust['Wins'] - Df5_3Clust['ExpWins'])
AvgError = sum(Df5_3Clust['Error'])/len(Df5_3Clust)
```

### 5 Clusters

```python
## 5 clusters
KM5_5 = KMeans(n_clusters = 5, random_state = 0, init = "k-means++", n_init = 100)
KM5_5.fit(X)
Df5_5Clust = Df5.copy()
Df5_5Clust['Cluster'] = KM5_5.labels_
KM5_5.cluster_centers_
```

```python
##Match Average Offense, Defense, Special spending back to groupby to gain cluster averages for all metrics
Df5_5_GB = Df5_5Clust.groupby(by = "Cluster")[['SuperBowl Win', "Playoff Wins",'Wins']].mean()
Df5_5_GB.reset_index(inplace = True)
labs = X.columns
for lab in labs:
    Df5_5_GB[lab] = 0

for i in range (0,len(KM5_5.cluster_centers_)):
    k_m = 0
    for lab in labs:
        val = KM5_5.cluster_centers_[i][k_m]
        Df5_5_GB.set_value(i,lab,val)
        k_m+=1
        
Df5_5_GB['TotCap'] = Df5_5_GB['Dbacks'] + Df5_5_GB['Front7'] + Df5_5_GB['Fantasy'] + Df5_5_GB['Oline'] + Df5_5_GB['Special']
Df5_5_GB['Count'] = Df5_5Clust.groupby(by = "Cluster")['Year'].count()
Df5_5_GB
```

```python
## Calculate Average Prediction Error
for i in range(0,5):
    Df5_3Clust[i] = 0

def DistFunc(x):
    D0 = np.sqrt((x['Dbacks']-Df5_5_GB.iat[0,4])**2+(x['Front7'] - Df5_5_GB.iat[0,5])**2 + (x['Fantasy'] - Df5_5_GB.iat[0,6])**2
                +(x['Oline']-Df5_5_GB.iat[0,7])**2+(x['Special']-Df5_5_GB.iat[0,8])**2)
    
    D1 = np.sqrt((x['Dbacks']-Df5_5_GB.iat[1,4])**2+(x['Front7'] - Df5_5_GB.iat[1,5])**2 + (x['Fantasy'] - Df5_5_GB.iat[1,6])**2
                +(x['Oline']-Df5_5_GB.iat[1,7])**2+(x['Special']-Df5_5_GB.iat[1,8])**2)
    
    D2 = np.sqrt((x['Dbacks']-Df5_5_GB.iat[2,4])**2+(x['Front7'] - Df5_5_GB.iat[2,5])**2 + (x['Fantasy'] - Df5_5_GB.iat[2,6])**2
                +(x['Oline']-Df5_5_GB.iat[2,7])**2+(x['Special']-Df5_5_GB.iat[2,8])**2)
    
    D3 = np.sqrt((x['Dbacks']-Df5_5_GB.iat[3,4])**2+(x['Front7'] - Df5_5_GB.iat[3,5])**2 + (x['Fantasy'] - Df5_5_GB.iat[3,6])**2
                +(x['Oline']-Df5_5_GB.iat[3,7])**2+(x['Special']-Df5_5_GB.iat[3,8])**2)
    
    D4 = np.sqrt((x['Dbacks']-Df5_5_GB.iat[4,4])**2+(x['Front7'] - Df5_5_GB.iat[4,5])**2 + (x['Fantasy'] - Df5_5_GB.iat[4,6])**2
                +(x['Oline']-Df5_5_GB.iat[4,7])**2+(x['Special']-Df5_5_GB.iat[4,8])**2)
    
    Tot = D0 + D1 + D2 + D3 + D4
    
    N0 = 1/((D0/Tot))
    N1 = 1/((D1/Tot))
    N2 = 1/((D2/Tot))
    N3 = 1/((D3/Tot))
    N4 = 1/((D4/Tot))
    TotN = N0 + N1 + N2 + N3 + N4
    
    Perc0 = N0/TotN*100
    Perc1 = N1/TotN*100
    Perc2 = N2/TotN*100
    Perc3 = N3/TotN*100
    Perc4 = N4/TotN*100
    
    return pd.Series([Perc0,Perc1,Perc2,Perc3,Perc4])

def ExpectedWins(x):
    WinShares0 = x[0]/100*Df5_5_GB.iat[0,3]
    WinShares1 = x[1]/100*Df5_5_GB.iat[1,3]
    WinShares2 = x[2]/100*Df5_5_GB.iat[2,3]
    WinShares3 = x[3]/100*Df5_5_GB.iat[3,3]
    WinShares4 = x[4]/100*Df5_5_GB.iat[4,3]
    ExpWins = WinShares0 + WinShares1 + WinShares2 + WinShares3 + WinShares4
    return ExpWins

Df5_5Clust[[0,1,2,3,4]] = Df5_5Clust.apply(DistFunc, axis = 1)
    
Df5_5Clust['ExpWins'] = Df5_5Clust.apply(ExpectedWins, axis = 1)
Df5_5Clust['Error'] = abs(Df5_5Clust['Wins'] - Df5_5Clust['ExpWins'])
AvgError = sum(Df5_5Clust['Error'])/len(Df5_5Clust)
```
