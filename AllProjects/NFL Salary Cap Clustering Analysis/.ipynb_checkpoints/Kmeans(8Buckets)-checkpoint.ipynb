```python
import pandas as pd
import sklearn as sk
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
Df3 = pd.read_csv('3Buckets.csv')
Df5 = pd.read_csv('5Buckets.csv')
Df8 = pd.read_csv('8Buckets.csv')
Df12 = pd.read_csv('12Buckets.csv')
```

```python
## Test if normalization by year affects the analysis
labels = ['Dbacks','LineBackers','Dline','PassCatchers','Oline','QB','RB','Special']
Df8Normalized = Df8.copy()

def Normalize(x):
    RetList = []
    for label in labels:
        dfyear = Df8Normalized[Df8Normalized['Year'] == x['Year']]
        val = (x[label] - np.mean(dfyear[label]))/(np.std(dfyear[label]))
        
        RetList.append(val)
    
    return pd.Series(RetList)
    
    
    
Df8Normalized[['Dbacks','LineBackers','Dline','PassCatchers','Oline','QB','RB','Special']] = Df8Normalized.apply(Normalize, axis = 1)   
```

```python
X = Df8[['Dbacks','LineBackers','Dline','PassCatchers','Oline','QB','RB','Special']]
NonPredictorX = Df8[['Year','Team']]
Y = Df8[['Wins', 'Playoff Wins', 'SuperBowl Win']]

X_Norm = Df8Normalized[['Dbacks','LineBackers','Dline','PassCatchers','Oline','QB','RB','Special']]
NonPredictorX_Norm = Df8Normalized[['Year','Team']]
Y_Norm = Df8Normalized[['Wins', 'Playoff Wins', 'SuperBowl Win']]
```

```python
## Elbow Plot
models = []
wcss = []
for i in range(1,35): 
    i = int(i)
    KM = KMeans(n_clusters = i, random_state = 0, init = "k-means++", n_init = 10)
    KM.fit(X)
    wcss.append(KM.inertia_)
    
plt.plot(range(1,35),wcss)
plt.title("Elbow Plot")
plt.xlabel("Num Clusters")
plt.ylabel("Within Cluster Sum of Squares")

plt.show()
print("Seemingly the best number of clusters is around 17, will try between 10-20")
```

### 6 Clusters (Not Normalized)

```python
## 3 clusters
KM8_6 = KMeans(n_clusters = 6, random_state = 0, init = "k-means++", n_init = 100)
KM8_6.fit(X)
Df8_6Clust = Df8.copy()
Df8_6Clust['Cluster'] = KM8_6.labels_
KM8_6.cluster_centers_
```

```python
##Match Averages spending back to groupby to gain cluster averages for all metrics
Df8_6_GB = Df8_6Clust.groupby(by = "Cluster")[['SuperBowl Win', "Playoff Wins",'Wins']].mean()
Df8_6_GB.reset_index(inplace = True)
labs = X.columns
# for lab in labs:
#     Df5_3_GB[lab] = 0

for i in range (0,len(KM8_6.cluster_centers_)):
    k_m = 0
    for lab in labs:
        val = KM8_6.cluster_centers_[i][k_m]
        Df8_6_GB.set_value(i,lab,val)
        k_m+=1
        
Df8_6_GB['TotCap'] = Df8_6_GB['Dbacks'] + Df8_6_GB['LineBackers'] + Df8_6_GB['Dline'] + Df8_6_GB['PassCatchers'] + Df8_6_GB['Oline'] + Df8_6_GB['QB'] + Df8_6_GB['RB'] + Df8_6_GB['Special']
Df8_6_GB['Count'] = Df8_6Clust.groupby(by = "Cluster")['Year'].count()
Df8_6_GB
```

```python
## Calculate Average Prediction Error
for i in range(0,3):
    Df8_6Clust[i] = 0

def DistFunc(x):
    D0 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[0,4])**2+(x['LineBackers'] - Df8_6_GB.iat[0,5])**2 + (x['Dline'] - Df8_6_GB.iat[0,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[0,7])**2+(x['Oline']-Df8_6_GB.iat[0,8])**2+(x['QB']-Df8_6_GB.iat[0,9])**2
                +(x['RB']-Df8_6_GB.iat[0,10])**2+(x['Special']-Df8_6_GB.iat[0,11])**2)
    
    D1 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[1,4])**2+(x['LineBackers'] - Df8_6_GB.iat[1,5])**2 + (x['Dline'] - Df8_6_GB.iat[1,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[1,7])**2+(x['Oline']-Df8_6_GB.iat[1,8])**2+(x['QB']-Df8_6_GB.iat[1,9])**2
                +(x['RB']-Df8_6_GB.iat[1,10])**2+(x['Special']-Df8_6_GB.iat[1,11])**2)
    
    D2 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[2,4])**2+(x['LineBackers'] - Df8_6_GB.iat[2,5])**2 + (x['Dline'] - Df8_6_GB.iat[2,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[2,7])**2+(x['Oline']-Df8_6_GB.iat[2,8])**2+(x['QB']-Df8_6_GB.iat[2,9])**2
                +(x['RB']-Df8_6_GB.iat[2,10])**2+(x['Special']-Df8_6_GB.iat[2,11])**2)
    
    D3 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[3,4])**2+(x['LineBackers'] - Df8_6_GB.iat[3,5])**2 + (x['Dline'] - Df8_6_GB.iat[3,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[3,7])**2+(x['Oline']-Df8_6_GB.iat[3,8])**2+(x['QB']-Df8_6_GB.iat[3,9])**2
                +(x['RB']-Df8_6_GB.iat[3,10])**2+(x['Special']-Df8_6_GB.iat[3,11])**2)
    
    D4 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[4,4])**2+(x['LineBackers'] - Df8_6_GB.iat[4,5])**2 + (x['Dline'] - Df8_6_GB.iat[4,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[4,7])**2+(x['Oline']-Df8_6_GB.iat[4,8])**2+(x['QB']-Df8_6_GB.iat[4,9])**2
                +(x['RB']-Df8_6_GB.iat[4,10])**2+(x['Special']-Df8_6_GB.iat[4,11])**2)
    
    D5 = np.sqrt((x['Dbacks']-Df8_6_GB.iat[5,4])**2+(x['LineBackers'] - Df8_6_GB.iat[5,5])**2 + (x['Dline'] - Df8_6_GB.iat[5,6])**2
                +(x['PassCatchers']-Df8_6_GB.iat[5,7])**2+(x['Oline']-Df8_6_GB.iat[5,8])**2+(x['QB']-Df8_6_GB.iat[5,9])**2
                +(x['RB']-Df8_6_GB.iat[5,10])**2+(x['Special']-Df8_6_GB.iat[5,11])**2)
    
    Tot = D0 + D1 + D2 + D3 + D4 + D5
    
    N0 = 1/((D0/Tot))
    N1 = 1/((D1/Tot))
    N2 = 1/((D2/Tot))
    N3 = 1/((D3/Tot))
    N4 = 1/((D4/Tot))
    N5 = 1/((D5/Tot))
    TotN = N0 + N1 + N2 + N3 + N4 + N5
    
    Perc0 = N0/TotN*100
    Perc1 = N1/TotN*100
    Perc2 = N2/TotN*100
    Perc3 = N3/TotN*100
    Perc4 = N4/TotN*100
    Perc5 = N5/TotN*100
    
    return pd.Series([Perc0,Perc1,Perc2,Perc3,Perc4,Perc5])

def ExpectedWins(x):
    WinShares0 = x[0]/100*Df8_6_GB.iat[0,3]
    WinShares1 = x[1]/100*Df8_6_GB.iat[1,3]
    WinShares2 = x[2]/100*Df8_6_GB.iat[2,3]
    WinShares3 = x[3]/100*Df8_6_GB.iat[3,3]
    WinShares4 = x[4]/100*Df8_6_GB.iat[4,3]
    WinShares5 = x[5]/100*Df8_6_GB.iat[5,3]
    ExpWins = WinShares0 + WinShares1 + WinShares2 + WinShares3 + WinShares4 + WinShares5
    return ExpWins

Df8_6Clust[[0,1,2,3,4,5]] = Df8_6Clust.apply(DistFunc, axis = 1)
    
Df8_6Clust['ExpWins'] = Df8_6Clust.apply(ExpectedWins, axis = 1)
Df8_6Clust['Error'] = abs(Df8_6Clust['Wins'] - Df8_6Clust['ExpWins'])
AvgError = sum(Df8_6Clust['Error'])/len(Df8_6Clust)
```

```python
## 12 clusters
KM8_20 = KMeans(n_clusters = 40, random_state = 0, init = "k-means++", n_init = 100)
KM8_20.fit(X_Norm)
Df8_20Clust = Df8.copy()
Df8_20Clust['Cluster'] = KM8_20.labels_
#KM8_20.cluster_centers_
```

```python
##Match Averages spending back to groupby to gain cluster averages for all metrics
Df8_20_GB = Df8_20Clust.groupby(by = "Cluster")[['SuperBowl Win', "Playoff Wins",'Wins']].mean()
Df8_20_GB.reset_index(inplace = True)
labs = X.columns
# for lab in labs:
#     Df5_3_GB[lab] = 0

for i in range (0,len(KM8_20.cluster_centers_)):
    k_m = 0
    for lab in labs:
        val = KM8_20.cluster_centers_[i][k_m]
        Df8_20_GB.set_value(i,lab,val)
        k_m+=1
        
Df8_20_GB['TotCap'] = Df8_20_GB['Dbacks'] + Df8_20_GB['LineBackers'] + Df8_20_GB['Dline'] + Df8_20_GB['PassCatchers'] + Df8_20_GB['Oline'] + Df8_20_GB['QB'] + Df8_20_GB['RB'] + Df8_20_GB['Special']
Df8_20_GB['Count'] = Df8_20Clust.groupby(by = "Cluster")['Year'].count()
#Df8_20_GB
```

```python
## Calculate Average Prediction Error
for i in range(0,40):
    Df8_20Clust[i] = 0

def DistFunc(x):
    Dlist = []
    TotD = 0
    for i in range(0,40):
        D = np.sqrt((x['Dbacks']-Df8_20_GB.iat[i,4])**2+(x['LineBackers'] - Df8_20_GB.iat[i,5])**2 + (x['Dline'] - Df8_20_GB.iat[i,6])**2
                +(x['PassCatchers']-Df8_20_GB.iat[i,7])**2+(x['Oline']-Df8_20_GB.iat[i,8])**2+(x['QB']-Df8_20_GB.iat[i,9])**2
                +(x['RB']-Df8_20_GB.iat[i,10])**2+(x['Special']-Df8_20_GB.iat[i,11])**2)
        Dlist.append(D)
        TotD += D

    Nlist = []
    TotN = 0
    for i in range(0,40):
        N = 1/((Dlist[i])/TotD)
        Nlist.append(N)
        TotN += N
        
        
    Perclist = []
    for i in range (0,40):
        Perc = Nlist[i]/TotN*100
        Perclist.append(Perc)

    return (pd.Series(Perclist))

def ExpectedWins(x):
    WinShares = 0
    for i in range(0,40):
        WinShares += x[i]/100*Df8_20_GB.iat[i,3]
    return WinShares

def ExpectedPlayoffWins(x):
    POWinShares = 0
    for i in range(0,40):
        POWinShares += x[i]/100*Df8_20_GB.iat[i,2]
    return POWinShares

Df8_20Clust.iloc[:,18:58] = Df8_20Clust.apply(DistFunc, axis = 1)
Df8_20Clust['ExpWins'] = Df8_20Clust.apply(ExpectedWins, axis = 1)
Df8_20Clust['ExpPOWins'] = Df8_20Clust.apply(ExpectedPlayoffWins, axis = 1)
Df8_20Clust['Error'] = abs(Df8_20Clust['Wins'] - Df8_20Clust['ExpWins'])
Df8_20Clust['POError'] = abs(Df8_20Clust['Playoff Wins'] - Df8_20Clust['ExpPOWins'])
AvgError = sum(Df8_20Clust['Error'])/len(Df8_20Clust)
AvgPOError = sum(Df8_20Clust['POError'])/len(Df8_20Clust)
```
