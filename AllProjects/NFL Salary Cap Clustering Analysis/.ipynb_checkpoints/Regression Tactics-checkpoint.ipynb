## Lets try normalizing

```python
## Test if normalization by year affects the analysis
labels = ['Safeties','Corners','InsideBackers','Edge','InteriorDLine','WRs','TEs','InteriorOLine','Tackles','QB','RB','Special']
Df12GBNormalized = DfGB.copy()

def Normalize(x):
    RetList = []
    for label in labels:
        dfyear = Df12GBNormalized[Df12GBNormalized['Year'] == x['Year']]
        val = (x[label] - np.mean(dfyear[label]))/(np.std(dfyear[label]))
        
        RetList.append(val)
    
    return pd.Series(RetList)
    
    
    
Df12GBNormalized[labels] = Df12GBNormalized.apply(Normalize, axis = 1)   
```

```python
Df12GBNormalized
```

```python
newlist = []
list1 = []
for i in (range(0,50)):
    j = str(i)
    newlist.append(j)
newlist = newlist + ['Team','IR','Team','Error','POError','ExpWins','ExpPOWins','Wins','Playoff Wins','SuperBowl Win']
X_Norm = Df12GBNormalized.drop(newlist, axis= 1)
Y_Norm = Df12GBNormalized[['Wins']]
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_Norm,Y_Norm,test_size=.2)
```

```python
Y_Train = y_train1.values[:,0]
Y_Test = y_test1.values[:,0]
reg = RandomForestRegressor(n_estimators = 1000, criterion = "mae")
reg.fit(X_train1, Y_Train)
Y_Pred = reg.predict(X_test1)
R2 = reg.score(X_test1, Y_Test)
```
