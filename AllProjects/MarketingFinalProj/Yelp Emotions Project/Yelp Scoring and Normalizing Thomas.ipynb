```python
from pandas import Series, DataFrame
import pandas as pd
%pylab inline

df_full = pd.read_json('yelp_academic_dataset_review_cleaned.json', lines=True)
df_full.head()

import numpy as np
import random
#np.random.seed(0)
random.seed(0)
sample = random.sample(population = arange(100000), k = 100)
df = df_full.iloc[sample]
df

```

```python
df['text'].map(len).sort_values()
```

```python
import re
def lower(s):
    return s.lower()

df['text_lower'] = df['text'].map(lower)

def remove_special_characters(s):
    ret = re.sub('[\n]',' ', s)
    return re.sub('[^a-zA-Z0-9\s]', '', ret)

df['text_lower_cleaned'] = df['text_lower'].map(remove_special_characters)

df.head()
```

```python
import sys
df_emotion = pd.read_csv('emotions_dataset.csv')
df_emotion['word'] = df_emotion['word'].map(lower).map(str.rstrip)
df_emotion = df_emotion.drop(labels = range(109))
df_emotion.head()
```

```python
import nltk
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
stop_words = list(ENGLISH_STOP_WORDS)
def in_stop_words(s):
    return s in stop_words
in_stop_words_mask = df_emotion['word'].map(in_stop_words)
df_emotion['word'] = df_emotion['word'][-in_stop_words_mask]
df_emotion = df_emotion[-df_emotion['word'].isna()]
df_emotion.head()
```
